```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
X,Y=make_circles(n_samples=1000,noise=.05,factor=.5)
plt.scatter(X[:,0],X[:,1],c=Y,cmap='jet')
plt.axis('square')
plt.xticks(fontsize=14),plt.yticks(fontsize=14)
plt.show()
```

```python
def f(X):
    return np.sqrt((X**2).sum(1))
```

```python
gridx,gridy=np.meshgrid(np.linspace(-1.2,1.2,num=100),np.linspace(-1.2,1.2,num=100))
gridvec = np.concatenate((np.reshape(gridx,(10000,1)),np.reshape(gridy,(10000,1))),1)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,cmap='jet')
plt.contour(gridx, gridy, np.reshape(f(gridvec),(100,100)),17, colors="gray", linestyles="solid")
plt.axis('square')
plt.xticks(fontsize=14),plt.yticks(fontsize=14)
plt.show()
```

```python
Z=f(X)
n,bins=np.histogram(Z,bins=50,range=(.2,1.2))
plt.hist(Z[Y==0],bins)
plt.hist(Z[Y==1],bins)
plt.xlim((.2,1.2))
plt.xticks(fontsize=14),plt.yticks(fontsize=14)
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,5))
YC=['C'+y for y in Y.astype(str)]
ax[0].scatter(X[:,0],X[:,1],c=YC)
ax[0].contour(gridx, gridy, np.reshape(f(gridvec),(100,100)),17, colors="gray", linestyles="solid")
ax[0].axis('tight')
ax[0].set_xticklabels([-1.,-1.,.5,0,.5,1.],fontsize=14),ax[0].set_yticklabels([-1.,-1.,.5,0,.5,1.],fontsize=14)
ax[1].tick_params(axis='both', which='major', labelsize=14)
ax[1].hist(Z[Y==0],bins)
ax[1].hist(Z[Y==1],bins)
ax[1].set_xlim((.2,1.2))
ax[1].set_xlabel('t',fontsize=14)
ax[1].axis('tight')
plt.show()
```
